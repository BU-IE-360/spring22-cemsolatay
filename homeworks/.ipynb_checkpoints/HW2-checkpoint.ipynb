Loading necessary packages.

```r
library(data.table)
library(zoo)
library(lubridate)
library(ggplot2)
library(forecast)
```

Reading the data, arranging column classes and names, and store it in a data.table object.

```r
df <- read.csv("../datasets/HW2/IE360_Spring22_HW2_data.csv", colClasses = c("character", rep("numeric", 10)), 
               col.names = c("Year_Quarter", "UGS", "RNUV", "NLPG", "PU", "PG", "NUGV", "NDGV", "GNPA", "GNPC", "GNP"))
dt <- data.table(df)
df <- NULL
```

Convert date information to yearqtr objects. Add year and quarter columns.

```r
dt[, Year_Quarter:=as.yearqtr(format(Year_Quarter), "%Y_Q%q")]
dt[, Year:=year(Year_Quarter)]
dt[, Quarter := as.factor(quarter(Year_Quarter))]
```

## 1. Plotting the Time Series of UGS

```r
ggplot(dt[Year < 2007], aes(x = Year_Quarter, y = UGS)) + geom_line(color="coral") + geom_point(color="coral3") +
    labs(title = "UGS vs. Time", x = "Year-Quarter", y = "UGS (in 1000 m3)") +
    scale_x_yearqtr(breaks = dt[, Year_Quarter], minor_breaks = dt[, Year_Quarter]) +
    theme(axis.text.x=element_text(angle=75,  hjust=1.4, vjust = 1.4)) +
    scale_y_continuous(breaks = 100000*seq(from = 7, to = 14, by = 1), minor_breaks = 100000*seq(from = 7,to = 14,by = .5))
```

The unleaded gasoline sales seem to be decreasing since 2000 towards the end of 2006. Also there is an obvious seasonal pattern repeated at each 4 data points which is reasonable since it is a quarterly dataset and a year has 4 quarters. Because UGS have a decreasing trend over time, the time series is not stationary with respect to its mean. However, the time series seems to be stationary with respect to its variance since the variance stays at similar levels as far as it is observable on the plot.

## 2. Autocorrelation Functions of UGS

```r
acf(dt[Year < 2007, UGS], 12)
```

Autocorrelation functions get significant values at lag 1 and 4, i.e, exceeding the boundaries. ACF value at lag 1 indicates the presence of trend in the time series while lag 4 refers to seasonality as mentioned before. There is a strong autocorrelation between the same quarter each year. Lags 2, 6, 10 are much lower than the other ones while staying in the boundaries, so we can say that the UGS values may be negatively affected by the UGS values at 2 quarters behind and ahead.

## 3. Introducing New Variables

Year and Quarter variables were already added to the dataset.

```r
dt[, Trend := 1:.N]
dt[, UGS_lag1 := shift(UGS, 1)]
dt[, UGS_lag2 := shift(UGS, 2)]
dt_train <- dt[Year < 2007]
```

```r
model1 <- lm(UGS~Trend+Quarter, dt_train)
summary(model1)
checkresiduals(model1)
```

```r
model1 <- lm(UGS~Quarter+PU+UGS_lag2, dt)
summary(model1)
checkresiduals(model1)
mean(model1$residuals)
```
