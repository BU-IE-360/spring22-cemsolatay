## 1. Introduction
Aim of this report is to fit a model to predict quarterly unleaded gasoline sales using the data of 2000-2006, and provide forecasts for 2007.
## 2. Data Preparation

Loading necessary packages.

```r
library(data.table)
library(zoo)
library(lubridate)
library(ggplot2)
library(forecast)
library(GGally)
library(MASS)
```

Reading the data, arranging column classes and names, and store it in a data.table object.

```r
df <- read.csv("../datasets/HW2/IE360_Spring22_HW2_data.csv", colClasses = c("character", rep("numeric", 10)), 
               col.names = c("Year_Quarter", "UGS", "RNUV", "NLPG", "PU", "PG", "NUGV", "NDGV", "GNPA", "GNPC", "GNP"))
dt <- data.table(df)
df <- NULL
```

Convert date information to yearqtr objects. Add year and quarter columns.

```r
dt[, Year_Quarter:=as.yearqtr(format(Year_Quarter), "%Y_Q%q")]
dt[, Year:=year(Year_Quarter)]
dt[, Quarter := as.factor(quarter(Year_Quarter))]
```

## 3. Plotting the Time Series of UGS

```r
ggplot(dt[Year < 2007], aes(x = Year_Quarter, y = UGS)) + geom_line(color="coral") + geom_point(color="coral3") +
    labs(title = "UGS vs. Time", x = "Year-Quarter", y = "UGS (in 1000 m3)") +
    scale_x_yearqtr(breaks = dt[, Year_Quarter], minor_breaks = dt[, Year_Quarter]) +
    theme(axis.text.x=element_text(angle=75,  hjust=1.4, vjust = 1.4)) +
    scale_y_continuous(breaks = 100000*seq(from = 7, to = 14, by = 1), minor_breaks = 100000*seq(from = 7,to = 14,by = .5))
```

The unleaded gasoline sales seem to be decreasing since 2000 towards the end of 2006. Also there is an obvious seasonal pattern repeated at each 4 data points which is reasonable since it is a quarterly dataset and a year has 4 quarters. Because UGS have a decreasing trend over time, the time series is not stationary with respect to its mean. However, the time series seems to be stationary with respect to its variance since the variance stays at similar levels as far as it is observable on the plot.

## 4. Autocorrelation Functions of UGS

```r
acf(dt[Year < 2007, UGS], 12)
```

Autocorrelation functions get significant values at lag 1 and 4, i.e, exceeding the boundaries. ACF value at lag 1 indicates the presence of trend in the time series while lag 4 refers to seasonality as mentioned before. There is a strong autocorrelation between the same quarter each year. Lags 2, 6, 10 are much lower than the other ones while staying in the boundaries, so we can say that the UGS values may be negatively affected by the UGS values at 2 quarters behind and ahead.

## 5. Model Building

### 5.1. Analyzing Independent Variables

Year and Quarter (as factors to represent seasonality correctly in the linear regression setting) variables were already added to the dataset.

Checking the correlation between variables to see which variables are explaining UGS better and which independent variables are already correlated with each other. 

```r
ggpairs(dt[Year < 2007, -c("Year", "Quarter", "Year_Quarter")])
```

There are significantly negative correlations between UGS and the numbers of vehicles by fuel type and the fuel prices. However, the numbers of vehicles by fuel type as well as the fuel prices are almost perfectly correlated with each other (i.e., NLPG, NUGV, NDGV are highly correlated and PU, PG are highly correlated). The negative correlation between the gas prices and sales seems logical and intuitive to me but I can't figure out the reason behind the negative correlation between the number of vehicles and UGS. Therefore, I am going to try to derive a variable that incorporate information in a more meaningful way. Also to increase the model interpretability, I will only include the PU variable which is the average price of unleaded gasoline. On the other hand, there is a significant positive correlation between GNPA and UGS. Finally, RNUV has very low correlations with the other variables. Now I will further investigate some of the variables.

```r
ggplot(dt, aes(x=Year_Quarter)) + geom_line(aes(y=GNPA)) +
    geom_line(aes(y=GNP), color="blue") + 
    geom_line(aes(y=GNPC), color="red") +
    labs(title = "GNP, GNPC, GNPA vs. Time", x = "Year-Quarter", y = "GNP, GNPA, GNPC (adjusted with an index)",
        subtitle = "Blue line shows GNP\nRed line shows GNPC\nBlack line shows GNPA") +
    scale_x_yearqtr(breaks = dt[, Year_Quarter], minor_breaks = dt[, Year_Quarter]) +
    theme(axis.text.x=element_text(angle=75,  hjust=1.4, vjust = 1.4))
```

Aggregate and sector based GNP values are showing a very similar seasonality to the UGS has. Therefore, I think that the positive correlation with UGS is the result of the seasonality effect that governs both of them. There is an increasing trend in GNP in contrast of UGS while they have positive correlation. Also the scatter plots of GNP variables versus UGS from the previous chart seems more like randomly distributed. We can say that the raw correlation values may be misleading. But we can try to incorporate these information to support the explaining ability of seasonality variables.

```r
ggplot(dt, aes(x=Year_Quarter)) + geom_line(aes(y=NUGV), color="darkgreen") +
    labs(title = "NUGV vs. Time", x = "Year-Quarter", y = "Number of unleaded gasoline using vehicles in the traffic") +
    scale_x_yearqtr(breaks = dt[, Year_Quarter], minor_breaks = dt[, Year_Quarter]) +
    theme(axis.text.x=element_text(angle=75,  hjust=1.4, vjust = 1.4))
```

We see that an increasing trend exists in the number of unleaded gasoling using vehicles in contrast to the unleaded gasoline sales. At least we can say that either the main consumption of unleaded gasoline doesn't result from the vehicles or the usage per vehicle dropped significantly. But the relationship is not interpretable and looks like a spurious correlation. I would expect a positive correlation between them as well as GNP values.
### 5.2. Defining Variables

```r
dt[, Trend := 1:.N] # Trend variable from 1 to N

# Lagged UGS values
dt[, UGS_lag1 := shift(UGS, 1)]
dt[, UGS_lag2 := shift(UGS, 2)]
dt[, UGS_lag3 := shift(UGS, 3)]
dt[, UGS_lag4 := shift(UGS, 4)]
dt[, UGS_lag5 := shift(UGS, 5)]

dt[, PU_lag1 := shift(PU, 1)]

dt[, New_NUGV := NUGV * RNUV]
dt[, Lost_NUGV := NUGV-shift(NUGV,1)-New_NUGV]
dt[, Vehicle_Ratio := NUGV/NDGV] # Not the exact ratio since the units are different
```

### 5.3. Building Regression Models
Building a model with just trend.

```r
model1 <- lm(UGS~Trend, dt)
summary(model1)
checkresiduals(model1)
mean(model1$residuals)
```

Adding seasonality to overcome high residual autocorrelations.

```r
model2 <- lm(UGS~Trend+Quarter, dt)
summary(model2)
checkresiduals(model2)
mean(model2$residuals)
```

The residual autocorrelations at lag 1 and 2 are relatively high with respect to others. We can try to include UGS values with lag 1 and 2.

```r
model3 <- lm(UGS~Trend+Quarter+UGS_lag1, dt)
summary(model3)
checkresiduals(model3)
mean(model3$residuals)
```

```r
model4 <- lm(UGS~Trend+Quarter+UGS_lag1+UGS_lag2, dt)
summary(model4)
checkresiduals(model4)
mean(model4$residuals)
```

```r
model5 <- lm(UGS~Trend+Quarter+UGS_lag2, dt)
summary(model5)
checkresiduals(model5)
mean(model5$residuals)
```

The best adjusted R-squared value is obtained with just UGS_lag2. But high autocorrelation at lag 7 made me think of including UGS_lag3.

```r
model6 <- lm(UGS~Trend+Quarter+UGS_lag2+UGS_lag3, dt)
summary(model6)
checkresiduals(model6)
mean(model6$residuals)
```

That didn't work well. We can add the other independent variables.

```r
model7 <- lm(UGS~Trend+Quarter+UGS_lag2+PU, dt)
summary(model7)
checkresiduals(model7)
mean(model7$residuals)
```

I tried to include each variable as well as 1 quarter lagged PU, number of new unleaded gasoline using vehicles, number of lost unleaded gasoline using vehicles; and PU gave the most improvement in the adjusted R-squared and the residuals. However now the trend variable doesn't seem to be significant based on its p value of t-test.

```r
model8 <- lm(UGS~Quarter+UGS_lag2+PU, dt)
summary(model8)
checkresiduals(model8)
mean(model8$residuals) 
```

The residuals obtained with trend were more looking like normally distributed. So, I will go with the model including the trend variable. Furthermore, I tried adding more and more variables to improve model performance, and it did. But, actually I don't want to do so. Because, having number of diesel gasoline using vehicles with positive coefficient, number of unleaded gasoline using vehicles with negative coefficient, variables derived just to increase performance without a causality for such a small dataset, leads to poor interpretability and spuriousity as I mentioned before. To illustrate it better, I will build a model by employing forward and backward selection based on AIC and obtain the best performing model.

```r
exaggerated_model <- lm(UGS ~.-Year_Quarter-Year-Lost_NUGV, data = dt)
step_model <- stepAIC(exaggerated_model, direction = "both", 
                      trace = FALSE)
summary(step_model)
```

Adjusted R-squared value obtained is jumped to 0.9911 but I think this model has no value in terms of interpretability. So I will assign the final model to model7 which was including trend, quarter of the year, unleaded gasoline price index, and 2 quarters lagged unleaded gasoline sales.
### 5.4. Final Model and Analysis

```r
final_model <- model7
```

```r
summary(final_model)
checkresiduals(final_model)
mean(final_model$residuals)
plot(final_model)
ggplot(dt[Year < 2007 & Trend > 2], aes(x=Year_Quarter)) + geom_line(aes(y=UGS), color="coral") + geom_line(aes(y=final_model$fitted.values), color="cyan") +
    geom_point(aes(y=UGS), color="coral") + geom_point(aes(y=final_model$fitted.values), color="cyan") +
    labs(title = "Actual - Fitted UGS vs. Time", x = "Year-Quarter", y = "UGS (in 1000 m3)") +
    scale_x_yearqtr(breaks = dt[, Year_Quarter], minor_breaks = dt[, Year_Quarter]) +
    theme(axis.text.x=element_text(angle=75,  hjust=1.4, vjust = 1.4)) +
    scale_y_continuous(breaks = 100000*seq(from = 7, to = 14, by = 1), minor_breaks = 100000*seq(from = 7,to = 14,by = .5))
```

Firstly, p-value associated with the F-statistic of the overall model is very small indicating that the model is meaningful. When we look at the p-values of each variable's t-statistic, we can say that quarterly seasonality is the most important variable. Also the other variables except trend came out to be significant. The case with the trend is mentioned before. Intercept covers the effect of Quarter1 and the rest coefficients are relational to Quarter1. Price and Trend have negative coefficients as expected from the visualizations. Adjusted R-squared of the model is 0.9464 which is an acceptable, nice value. Residuals are ranging from -61 thousand to 52 thousand. The mean of the residuals is very close to 0 which is the assumption of linear regression. However, at the last year, the residuals seem to be getting larger and negatively autocorrelated. But it is a very small dataset and it is a very short time window to be worried. The histogram and the normal q-q plot of the residuals are suggesting that the residuals are distributed normally which is again an assumption of linear regression. Also, all of the autocorrelation function values at each lag is inside the boundaries which suggests the residuals are independent. The constant variance of the residuals assumption is holding if we ignore the last part.
### 5.5. Forecasting

```r
forecast <- c()
for(i in 1:4){
    forecast[i] <- predict(final_model, newdata = dt[Year == 2007 & Quarter == as.factor(i)]) 
    if(i <= 2){
        dt[Year == 2007 & Quarter == i+2]$UGS_lag2 <- forecast[i]
    }
}
dt[, Forecast := c(rep(NA, 2), final_model$fitted.values, forecast)]
dt[Year == 2007, c("Year_Quarter", "Forecast")]
```
